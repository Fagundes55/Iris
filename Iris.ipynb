```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length',hue='species',data=iris)
plt.grid()
plt.show()
```

```python
fig = px.scatter(iris, x='sepal_length', y='petal_length',color='species')

fig.update_layout(
    title = 'Iris',
    xaxis_title = 'Comprimento da Sépala',
    yaxis_title = 'Comprimento da Pétala'
    )
```

```python
plt.hist(iris['sepal_length'],edgecolor='white')
plt.title('Comprimento da sépala')
plt.show()
```

```python
sns.histplot(x='sepal_width',data=iris)
plt.show()
```

```python
px.histogram(iris, x='petal_length',cumulative=True,
             title='Largura da pétala')
```

```python
px.histogram(iris, x='petal_width', marginal='box',
             title='Largura da pétala')
```

```python
fig,ax = plt.subplots(2,2, constrained_layout=True,)
ax[0,0].hist(iris['petal_length'])
ax[0,0].set_title('Comprimento da pétala')
ax[0,0].set_ylabel('Frequência')

ax[0,1].hist(iris['petal_width'])
ax[0,1].set_title('Largura da pétala')

ax[1,0].hist(iris['sepal_length'])
ax[1,0].set_title('Comprimento da sépala')
ax[1,0].set_ylabel('Frequência')

ax[1,1].hist(iris['sepal_width'])
ax[1,1].set_title('Largura da Sépala')
plt.show()
```
